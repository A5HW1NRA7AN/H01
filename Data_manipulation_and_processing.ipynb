```python
import os
import pandas as pd

def find_common_columns_in_merged(directory, prefix='merged_', suffix='.csv'):
    """
    Finds common columns in all merged_*.csv files in a given directory.

    Parameters:
        directory (str): Path to directory containing merged CSVs.
        prefix (str): Filename prefix to filter relevant CSVs.
        suffix (str): File extension (usually .csv).

    Returns:
        set: Common column names across all merged files.
    """
    files = [f for f in os.listdir(directory) if f.startswith(prefix) and f.endswith(suffix)]
    common_cols = None

    for file in sorted(files):
        file_path = os.path.join(directory, file)
        try:
            df = pd.read_csv(file_path, nrows=5)  # Only read top 5 rows for speed
            columns = set(df.columns)

            if common_cols is None:
                common_cols = columns
            else:
                common_cols = common_cols.intersection(columns)

            print(f"✅ Loaded {file} with {len(columns)} columns")

        except Exception as e:
            print(f"❌ Failed to read {file}: {e}")

    print(f"\n🔍 Common columns across {len(files)} merged files:")
    for col in sorted(common_cols):
        print(f" - {col}")

    return common_cols
```

```python
import os
import pandas as pd

def merge_merged_csvs_on_common_columns(directory, output_filename='final_merged_dataframe.csv'):
    """
    Merges all merged_*.csv files using only common columns and saves the final DataFrame.

    Parameters:
        directory (str): Path to directory containing merged CSVs.
        output_filename (str): Name of the output CSV file.
    """
    # Step 1: Get list of files
    files = [f for f in os.listdir(directory) if f.startswith('merged_') and f.endswith('.csv')]
    files.sort()  # Optional: sort by year

    # Step 2: Find common columns
    common_cols = None
    for file in files:
        try:
            df = pd.read_csv(os.path.join(directory, file), nrows=5)
            cols = set(df.columns)
            common_cols = cols if common_cols is None else common_cols.intersection(cols)
        except Exception as e:
            print(f"❌ Failed to read {file}: {e}")

    if not common_cols:
        print("❌ No common columns found.")
        return

    print(f"\n✅ Common columns ({len(common_cols)}): {sorted(common_cols)}")

    # Step 3: Load full data using only common columns
    final_df_list = []
    for file in files:
        try:
            df = pd.read_csv(os.path.join(directory, file), usecols=common_cols)
            final_df_list.append(df)
            print(f"✅ Appended {file} with shape {df.shape}")
        except Exception as e:
            print(f"❌ Failed to load {file} with common columns: {e}")

    # Step 4: Concatenate all DataFrames
    final_df = pd.concat(final_df_list, ignore_index=True)
    output_path = os.path.join(directory, output_filename)
    final_df.to_csv(output_path, index=False)

    print(f"\n🎉 Final merged DataFrame shape: {final_df.shape}")
    print(f"💾 Saved to: {output_path}")
```

```python
# Define function to select independent features using correlation + mutual information
def select_features(X, y, top_k=15):
    correlations = X.corrwith(y.mean(axis=1)).abs()
    mutual_info = mutual_info_regression(X, y.mean(axis=1))
    scores = (correlations + mutual_info) / 2
    selected_features = scores.sort_values(ascending=False).head(top_k).index.tolist()
    return selected_features

# Select inputs and targets for Healthcare Accessibility
y_healthcare = data[healthcare_targets].dropna()
X_all = data.drop(columns=healthcare_targets)
X_healthcare = X_all.loc[y_healthcare.index]
selected_features = select_features(X_healthcare, y_healthcare)
X_healthcare = X_healthcare[selected_features]

# Normalize inputs
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X_healthcare)
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y_healthcare)

# Split the data
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Dataset class
class SocioHealthDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = SocioHealthDataset(X_train, y_train)
val_dataset = SocioHealthDataset(X_val, y_val)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)

# Model with LSTM + MLP for long-term dependencies
class HealthcareModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(HealthcareModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, 64, batch_first=True)
        self.fc1 = nn.Linear(64, 128)
        self.fc2 = nn.Linear(128, output_dim)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = x.unsqueeze(1)  # LSTM expects (batch, seq_len, input_size)
        out, _ = self.lstm(x)
        x = self.relu(self.fc1(out[:, -1, :]))
        x = self.fc2(x)
        return x

# Initialize model, loss, optimizer
model = HealthcareModel(input_dim=X_train.shape[1], output_dim=y_train.shape[1])
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# TensorBoard writer
writer = SummaryWriter("runs/healthcare_model")

# Train function with early stopping

def train_model(model, train_loader, val_loader, criterion, optimizer, epochs=50, device='cpu', patience=5):
    model.to(device)
    best_loss = float('inf')
    patience_counter = 0

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for features, targets in train_loader:
            features, targets = features.to(device), targets.to(device)

            optimizer.zero_grad()
            outputs = model(features)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        avg_train_loss = running_loss / len(train_loader)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for features, targets in val_loader:
                features, targets = features.to(device), targets.to(device)
                outputs = model(features)
                loss = criterion(outputs, targets)
                val_loss += loss.item()

        avg_val_loss = val_loss / len(val_loader)

        writer.add_scalars('Loss', {
            'Train': avg_train_loss,
            'Validation': avg_val_loss
        }, epoch)

        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

        # Early stopping
        if avg_val_loss < best_loss:
            best_loss = avg_val_loss
            patience_counter = 0
            torch.save(model.state_dict(), 'best_healthcare_model.pth')
            torch.save(model.state_dict(), 'best_healthcare_model.h5')
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print("Early stopping triggered")
                break

    print("Training complete.")

train_model(model, train_loader, val_loader, criterion, optimizer, device='cuda' if torch.cuda.is_available() else 'cpu')
```
