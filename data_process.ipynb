```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

import pandas as pd

class DataProcessingAPI:
    def __init__(self, df):
        """
        Initialize the DataProcessingAPI class and classify numeric and categorical columns.

        Parameters:
        df (pd.DataFrame): Input dataframe to classify columns.
        """
        self.df = df.copy()  # Use a copy to avoid modifying the original dataframe
        self.numeric_cols = self.df.select_dtypes(include=['float64', 'int64']).columns.tolist()
        self.categorical_cols = self.df.select_dtypes(include=['object']).columns.tolist()

    def process(self):
        # Drop columns with >= 50% missing values
        missing_columns = self.df.columns[self.df.isnull().mean() >= 0.5]
        self.df.drop(columns=missing_columns.tolist(), inplace=True)

        # Data imputation
        for col in self.df.columns:
            if col in self.categorical_cols:
                mode_val = self.df[col].mode()[0]
                self.df[col] = self.df[col].fillna(mode_val)
            else:
                skew_val = self.df[col].skew()
                if skew_val < 0.5:
                    mean_val = self.df[col].mean()
                    self.df[col] = self.df[col].fillna(mean_val)
                else:
                    median_val = self.df[col].median()
                    self.df[col] = self.df[col].fillna(median_val)

        return self.df
```

```python
import os
import pandas as pd

# Define the input and output directories
input_dir = r"D:\hackToFuture\data\County (2011-20)"
output_dir = r"D:\hackToFuture\new_data1\County (2011-20)"

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Process each Excel file in the input directory
for file_name in os.listdir(input_dir):
    if file_name.endswith(".xlsx"):
        # Read the Excel file
        file_path = os.path.join(input_dir, file_name)
        df = pd.read_excel(file_path, sheet_name='Data')
        
        data_processing_pipeline = DataProcessingAPI(df)

        # Process the data using the pipeline
        processed_df = data_processing_pipeline.process()
        
        # Save the processed data as a CSV file
        output_file_path = os.path.join(output_dir, file_name.replace(".xlsx", ".csv"))
        processed_df.to_csv(output_file_path, index=False)
```
